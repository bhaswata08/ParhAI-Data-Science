# Engineering systems for data science

**Systems thinking**: Systems thinking is away of thinking that is global and encompassing rather than focussed on particular issues Eg: Quicksort instead of bubblesort





# CRISP-DM

### Process



### CRISP-DM
1. Business understanding
2. Data Understanding
3. Data Preparation
4. Modelling
5. Evaluation
6. Deployment





### Business Understanding
1. What are my business objectives(Captured in a systemetic way)?
2. Can Data Science actually achieve those objectives?
 > * Data science can sometimes do nothing else other than reveal the existing problems in organization which need to be solved elsewhere
3. How do you define success metrics?
4. Are there ethical usuage in data usuage?
5. What have other industries achieved?

### Data Understanding
1. What are the souces of data
2. Does new daa need to be collected
3. What is the quantity and quality of data available?
4. What do the differnt data items represent?
5. Which data is relevent to the objectives?

### Data Preparation
1. What are the different data formats?
2. is there need for annotationg data?
3. How can data be extracted, transformed and loaded?
4. How to standardize and normalize data?
5. How to efficiently store data for analysis?


### Data Modelling
1. What assumptions to make for the models?
2. Statistical or algorithmic modelling?
3. is the cleandata we have sufficient for modelling or do we need more data?
4. Is the compute budget sufficient for modelling?
5. Are results statistically significant?


### Evaluation
1. Does model work correctly on test data?
2. Does model achieve business objectives?
3. Does model meet performance requirements?
4. Is the model unbiased and robust?
5. What are the ways to improve the model

### Deployment
1. Where is the model to be deployed
2. What is the Hardware/Software stack for deployment?
3. Does it meet performance requirements
> * Eg. Limited battery on phone, very low latency for stock trading
4. Does it violate privacy requirements?
5. Does it meet users' expectations

### Iterative design
1. Iterative design and deployment(Minimum viable product(MVP))
> MVP: Instead of building very complex things, build a simple thing and iterate
2. Revise expectations of success and value of data science
3. Upgrade human and hardware resources
