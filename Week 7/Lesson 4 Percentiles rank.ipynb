# How do you compute percentile rank of a value in the data?




 Compared to other students, how do you rate the performance of the student who scored 44?

OR

 What is the percentile rank of the student who scored 44?

OR

 If I had the data in sorted order, And if I look at the element 44, What percentage of the values would lie below it?

* The percentile rank of a value is the percentage of values in the data set that are less than or equal to it.

EXAMPLE 1:



>$PR_s = \frac{c_s+0.5*f_s}{n}*100$

$PR_s$ = percentile rank of score s

$c_s$ = number of values less than s

$f_s$ = number of values equal to s

$PR_{44} = \frac{6+0.5 * 2}{25} * 100$ = 28

How should you interpret this percentile: This person has not performed very well in this course because only 28 % of the students have scored less than him and 72% of the students have scored more than him.

EXAMPLE 2:

Shikhar Dhawan T20I scores (59 sorted scores)



> $PR_s = \frac{c_s+0.5*f_s}{n}*100$
>
> $PR_{32}$ = $\frac{37+0.5 * 2}{59} * 100$ = 64.40$

* We typically round off the percentile rank to the nearest whole number. (65 in this case)
  
How should you interpret this percentile: 65% of the times Shikhar Dhawan has scored less than 32 runs in T20I.


# Effect of Transformation on Percentiles

## Scaling and shifting



$x_{new} = a * x +c (a= \frac{5}{9}, c = \frac{-160}{9})$

## Effect of transformations on percentiles

>$x_{new} = a * x +c$
>
>${L}_{p}^{new} = \frac{p}{100} * (n+1) = L_p$

Location of pth percentile remains unchanged, because the relative ordering of the data doesn't change.

>${i}_{p}^{new} = i_p, {f}_{p}^{new} = f_p$
>
>$Y_{p}^{new} = {x}_{i_p}^{new} + f_p * ({x}_{i_{p+1}}^{new}-{x}_{i_p}^{new})$

Substituting for $x_{i_p}^{new}$ and $x_{i_{p+1}}^{new}$:

>${Y}_{p}^{new} = \mathrm{x}_{i_p}^{new} + f_p * (\mathrm{x}_{i_{p+1}}^{new}-\mathrm{x}_{i_p}^{new})$
>
>$Y_p^{new} = (a*x_{i_p}+c) + f_p * (a*x_{i_{p+1}}+\bcancel{c} - (a*x_{i_p}+\bcancel{c}))$
>
>$\;\;\;\;\;\;\;\;\: = a*x_{i_p} + c + f_p * (a*x_{i_{p+1}} - a*x_{i_p})$
>>
>$\;\;\;\;\;\;\;\;\:= a* (x_{i_p} + f_p * (x_{i_{p+1}} - x_{i_p})) + c$
>>
>$\;\;\;\;\;\;\;\;\:= a * Y_p +c$

* The new value of the pth percentile is the same as the old value of the pth percentile, multiplied by a and added to c.
