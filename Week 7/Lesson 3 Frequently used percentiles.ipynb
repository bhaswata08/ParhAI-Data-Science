# Frequently used percentiles

## Quartiles



Quartiles divide the data into four equal parts. The first quartile (Q1) is the 25th percentile, the second quartile (Q2) is the 50th percentile, and the third quartile (Q3) is the 75th percentile.

EXAMPLE: 

Shikhar Dhawan T20I scores (50 sorted scores)




> <span style="color:yellow"> $L_{25} = \frac{25}{100} * (50+1) = 12.75$ </span>
>
> <span style="color:yellow"> $Q_1 = Y_{25} = x_{12} + 0.75 * (x_{13} - x_{12}) = 5$ </span>


> <span style="color:DodgerBlue"> $L_{50} = \frac{50}{100} * (50+1) = 25.5$ </span>
>
> <span style="color:DodgerBlue"> $Q_2 = Y_{50} = x_{15} + 0.5 * (x_{16} - x_{15}) = 15.5$ </span>


> <span style="color:Orange"> $L_{75} = \frac{75}{100} * (50+1) = 38.25$ </span>
>
> <span style="color:Orange"> $Q_3 = Y_{75} = x_{38} + 0.25 * (x_{39} - x_{38}) = 42.25$ </span>


* Median is same as $Q_2$



* But why do the formula look so different
  
<span style="color:DodgerBlue"> $L_{50} = \frac{50}{100} * (50+1) = 25.5$ </span>

CASE 1: If the n is odd

$L_{50}$ will be an integer ( n+1 is even)

$i_p = L_p = \frac{n+1}{2}, f_p = 0$

<span style="color:DodgerBlue"> $Q_2 = Y_{50} = x_{i_p} + 0 * (x_{i_p + 1} - x_{i_p}) = x_{\frac{x+1}{2}}$ </span>

CASE 2: If the n is even

$L_{50}$ will be a fraction ( n+1 is odd)

$i_p = L_p = \frac{n+1}{2} = \frac{n}{2} + \frac{1}{2} = i_p +0.5$

$i_p = \frac{n}{2}, f_p = 0.5 = \frac{1}{2}$

<span style="color:DodgerBlue"> $Q_2 = Y_{50} = x_{i_p} + 0.5 * (x_{i_p + 1} - x_{i_p}) = \frac{x_{\frac{n}{2}} + x_{\frac{n}{2} + 1}}{2}$ </span>

## Quintiles



Quintiles divide the data into five equal parts. The first quintile (P1) is the 20th percentile, the second quintile (P2) is the 40th percentile, the third quintile (P3) is the 60th percentile, the fourth quintile (P4) is the 80th percentile, and the fifth quintile (P5) is the 100th percentile.

EXAMPLE: Shikhar Dhawan T20I scores (50 sorted scores)



TO FIND: P3

$L_60 = \frac{60}{100} * (50+1) = 30.6$

$P_3 = Y_{60} = x_{30} + 0.6 * (x_{31} - x_{30}) = 27.8$

Similarly we can find P1, P2, P4, P5

## Deciles



* Deciles divide the data into ten equal parts. The first decile (D1) is the 10th percentile, the second decile (D2) is the 20th percentile, the third decile (D3) is the 30th percentile, the fourth decile (D4) is the 40th percentile, the fifth decile (D5) is the 50th percentile, the sixth decile (D6) is the 60th percentile, the seventh decile (D7) is the 70th percentile, the eighth decile (D8) is the 80th percentile, the ninth decile (D9) is the 90th percentile, and the tenth decile (D10) is the 100th percentile.

EXAMPLE: Shikhar Dhawan T20I scores (50 sorted scores)



TO FIND: D3

$L_30 = \frac{30}{100} * (50+1) = 15.3$

$D_3 = Y_{30} = x_{15} + 0.3 * (x_{16} - x_{15}) = 5.3$

Similarly we can find D1, D2, D4, D5, D6, D7, D8, D9, D10
