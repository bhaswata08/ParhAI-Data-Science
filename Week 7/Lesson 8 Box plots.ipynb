# Box Plots

* Box plots are used for visualizing spread, median and outliers in the data.
  
________________________________________________________________________________________________
  
**Outliers:**: We say x is an outlier if
> x < Q1 - 1.5 * IQR 
>
> or 
>
> x > Q3 + 1.5 * IQR

*Where Q1 and Q3 are the first and third quartiles and IQR is the interquartile range.*
_________________________________________________________________________________________________



EXAMPLE: Shikhar Dhawan T20I scores (59 sorted scores)



* Median = 23
* Q1 = 5, Q3 = 41
* IQR = 36
* Q1 - 1.5 * IQR = -49
* Q3 + 1.5 * IQR = 95

No outliers in this case.



### Box plot(variant 1)

* Median = 32
* Q1 = 27.75, Q3 = 38.25
* IQR = 10.5
* Q1 - 1.5 * IQR = 12
* Q3 + 1.5 * IQR = 54
* min = 2, max = 45



* In case of data having no outliers, the whiskers are drawn to the min and max values.



### Box plot(variant 2)

* Median = 32
* Q1 = 27.75, Q3 = 38.25
* IQR = 10.5
* Q1 - 1.5 * IQR = 12
* Q3 + 1.5 * IQR = 54
* min = 2, max = 45



* Five number summary: min, Q1, median, Q3, max (in this order)
* Not popular because it doesn't show the outliers in the data

### Box plot(variant 3)

* Median = 32
* Q1 = 27.75, Q3 = 38.25
* IQR = 10.5
* Q1 - 1.5 * IQR = 12
* Q3 + 1.5 * IQR = 54
* min = 16 (Excluding outliers), max = 45 (Excluding outliers)



* Popular because it shows the outliers in the data and features of the data itself

## Boxplots: (Usage in ML)





Ideal Behavior: No overlap in the IQR of the scores assigned to positive and negative reviews.

### Box plots to visually compare the performance of ML systems:



Model 1: Not a good model because it has a lot of overlap in the IQR of the scores assigned to positive and negative reviews.

Model 2: Not a good model because there are some positive reviews which have been assigned a negative score, and vice versa.

Model 3: A good model because there is no overlap in the IQR of the scores assigned to positive and negative reviews. It can clearly differentiate between positive and negative reviews.
