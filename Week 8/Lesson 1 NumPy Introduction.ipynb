# Numpy

## Working with Arrays



* All of these require storing and processing high dimensional arrays efficiently

BUT

We already have lists, sets, tuples and dictonaries

> Lists can store collection of high dimnesional numbers as arrays and we can operate on them by iterating over them

But in reality, this is very inefficient - 10x to 100x slower than expected performance

❓ But why is this inefficient?

* Lists are designed to store heterogeneous data
* No low-level hardware mechanisms to accelerate operations on lists



Intended to bring performance and functionality improvements for numerical computing

* Numpy is a Python library that provides a multidimensional array object, various derived objects (such as masked arrays and matrices), and an assortment of routines for fast operations on arrays, including mathematical, logical, shape manipulation, sorting, selecting, I/O, discrete Fourier transforms, basic linear algebra, basic statistical operations, random simulation and much more.

 | **What NumPy does?**                             |                                                                                                                                                                        |
|--------------------------------------------------|------------------------------------------------------------------------------------------------------------------------------------------------------------------------|
| At the user perspective                          | * Provides implementations of many functions across linear algebra, stats... * Efficiently broadcast operations across dimensions                                      |
| At the programming/functionality language level  | * Enables other packages to use numpy arrays as an efficient data interface * Efficiently processes data without the type-checking overhead(i.e only homogeneous data) |
| As an interface to hardware and operating system | * Enable easy and efficient file save and load of n-d arrays * Efficiently store n-d arrays in a vectorized form to benefit from the DRAM locality                     |
