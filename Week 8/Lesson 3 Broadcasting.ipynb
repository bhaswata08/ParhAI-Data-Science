Broadcasting: We replicate values of one or more operands to be expanded along other dimensions to match the shape of the other operand.

BUT

 When is broadcasting applicable?

* The dimensions of the two operands must be compatible.
* The dimension of the two operands must be equal or one of them must be 1.
* Python starts backwards from the last dimension and compares the dimensions of the two operands.
* If the dimensions are equal or one of them is 1, the dimensions are compatible.
* If the dimensions are not equal and neither of them is 1, the dimensions are not compatible.
* If dimensions are missing, they are assumed to be 1.
